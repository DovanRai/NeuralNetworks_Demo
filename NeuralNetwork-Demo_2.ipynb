```python
#Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
#load dataset

(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data() #train-test 0.75-0.25 split
```

```python
# summarize loaded dataset
print('Train: X=%s, y=%s' % (x_train.shape, y_train.shape))
print('Test: X=%s, y=%s' % (x_test.shape, y_test.shape))
```

```python
#normalize data by scaling

x_train = x_train / 255
x_test = x_test / 255
```

```python
#flatten x grids into longitutdal input

image_vector_size = 28*28

x_train = x_train.reshape(len(x_train), image_vector_size)
x_test = x_test.reshape(len(x_test), image_vector_size)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define Neural Network model architecture with one hidden layer with 32 neurons

image_size = 784 # 28*28
num_classes = 10 # ten unique digits

model = Sequential()
model.add(Dense(units=32, activation='sigmoid', input_shape=(image_size,)))
model.add(Dense(units=num_classes, activation='softmax'))
model.summary()
```

```python
#compile and train the neural network model

model.compile(optimizer="sgd", loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history= model.fit(x_train, y_train, batch_size=100, epochs=50, validation_split=0.2, verbose=0)

# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy- one hidden layer with 32 neurons')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()
```

```python
# Define Neural Network model architecture with two hidden layers with 32 neurons in each

image_size = 784 # 28*28
num_classes = 10 # ten unique digits

model = Sequential()
model.add(Dense(32, activation='sigmoid', input_shape=(image_size,)))
model.add(Dense(32, activation='sigmoid'))
model.add(Dense(units=num_classes, activation='softmax'))
model.summary()

#compile and train the neural network model

model.compile(optimizer="sgd", loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history= model.fit(x_train, y_train, batch_size=100, epochs=50, validation_split=0.2, verbose=0)

# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy- two hidden layers with 32 neurons in each')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()
```
